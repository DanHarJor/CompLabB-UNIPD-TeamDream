```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 14

import keras
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten #, Reshape
from keras.layers import Conv1D, MaxPooling1D, AveragePooling1D

import numpy as np
import matplotlib.pyplot as plt

def Show_data(x,L,nSamples = 2,title="data"):
    plt.figure(figsize = (nSamples*3,10))
    for i in range(nSamples):
        plt.plot(np.arange((i+1)*L,(i+1)*L+L),x[i])
        #plt.plot(np.arange(L,2*L),x[1])
        #plt.plot(np.arange(2*L,3*L),x[2])
    plt.title(title)
    plt.xlabel("time")
    plt.show()
```

### Read data

```python
str0 = 'ts_L60_Z12_A500_DX50_bias5_N10000.dat'
fnamex='DATA/x_'+str0
fnamey='DATA/y_'+str0

x = np.loadtxt(fnamex, delimiter=" ",dtype=float)
N,L = len(x), len(x[0])

Show_data(x,L,6,"original data")

categ = np.loadtxt(fnamey, dtype=int)
n_class = 3    # y.argmax() - y.argmin() +1
print('data: ',N)

y = np.zeros((N,n_class))
for i in range(N):
    y[i][categ[i]] = 1.
```

### Rescale data, split train/val.

```python
import numpy as np
import matplotlib.pyplot as plt

def Show_data(x,L,nSamples = 2,title="data"):
    fig = plt.figure(figsize = (nSamples*3,10))
    for i in range(nSamples):
        plt.plot(np.arange((i+1)*L,(i+1)*L+L),x[i])
        #plt.plot(np.arange(L,2*L),x[1])
        #plt.plot(np.arange(2*L,3*L),x[2])
    plt.title(title)
    plt.xlabel("time")
    plt.show()

#remove average value of each sample from its values
xm = x.mean(axis=1)
for i in range(N):
    x[i] = x[i]-xm[i]

#rescale crude
x = x/400

Show_data(x,L,6,"rescaled data")

perc_train=0.8
N_train = int(perc_train*N)
x_train = x[:N_train]
y_train = y[:N_train]
x_val = x[N_train:]
y_val = y[N_train:]
N_val = len(x_val)
print('N_train=',N_train,'  N_val=',N_val,'  L=',L,'  n_class=',n_class)

# Keras wants an additional dimension with a 1 at the end
x_train = x_train.reshape(x_train.shape[0], L, 1)
x_val =  x_val.reshape(x_val.shape[0], L, 1)
input_shape = (L, 1)
```

## Convolutional NN

```python
def Show_weights(model,l=0,label="model"):
    c=['r','y','c','b','m']
    m=['o','s','D','<','>']
    ms=10
    
    w = model.layers[l].get_weights()[0]
    wT=w.T
    M=len(wT)
    b = model.layers[l].get_weights()[1]
    
    fig,AX=plt.subplots(1,2,figsize=(12,4.4))
    ax=AX[0]
    ax.axhline(0, c="k")
    ax.plot((0,))
    for i in range(M):
        ax.plot(wT[i][0],"-",c=c[i],marker=m[i],label=str(i),markersize=ms)
    ax.set_title(label+': filters of layer '+str(l))
    ax.set_xlabel('index')
    ax=AX[1]
    ax.axhline(0, c="k")
    for i in range(M):
        ax.plot((i),(b[i]),c=c[i],marker=m[i],label="filter "+str(i),markersize=ms)
    ax.set_title(label+': bias of layer '+str(l))
    ax.set_xlabel('filter nr')
    ax.set_xticks(np.arange(5))
    ax.legend()
    plt.show()
    
Show_weights(model,0)
Show_weights(model,2)
```

# Daniel's Regularisation Investigation

### Determining the regularisation parameter's direction

When regularisation is included the aim of gradient descent is to minamise both the cost function and the regularisation function. The regularisation parameter $\lambda$ or $D$ tunes how much effect the regularisation has. In keras the parameter is inserted here: tf.keras.regularizers.l1(regLevel). I am unsure wether regLevel is $\lambda$ or $D$. They are the inverse of each other. With more $\lambda$ (or less $D$) we have more regularisation. 

argmin(Cost + $\lambda$ Regularisation)

argmin($D$ Cost + Regularisation)

Regularisation restricts over-fitting so with more regularisation I should get less-overfitting. The model will be trained with 10 different regLevels from 0 to 6. If overfitting decreases then regLevel = $\lambda$. If overfitting increases then reglevel = $D$. 

```python
import os
# ...LESSON ...
# Hyper-parameters
# with small minibatch it does not converge!! 


BATCH_SIZE = 250
EPOCHS = 100

#used to retain best weights when training
model_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath = os.getcwd(),
    save_weights_only=False,
    monitor='val_accuracy',
    mode='max',
    save_best_only=True,
    verbose = 0)

accTrain = []
accVal = []
regLevels = np.linspace(0,6,10)#np.array([0.1,0.001,1e-5,0])#
for regLevel in regLevels:
    compile_model(regLevel)
    fit = model.fit(x_train,y_train,batch_size=BATCH_SIZE,
                epochs=EPOCHS,
                validation_data=(x_val, y_val),
                verbose=0, shuffle=True,callbacks=[model_checkpoint_callback])
    accTrain.append(np.max(fit.history['accuracy']))
    accVal.append(np.max(fit.history['val_accuracy']))
```

```python
regFig = plt.figure(figsize = (15,7))
plt.plot(regLevels,accTrain, label = 'Training Accuracy, Tac')
plt.plot(regLevels,accVal, label = 'Validation Accuracy, Vac')

plt.xlabel('regLevel')
plt.ylabel('Accuracy')
plt.legend()

overFig = plt.figure(figsize = (15,7))
plt.plot(regLevels, np.array(accTrain)-np.array(accVal), label = 'Overfitting Level, Tac - Vac')
plt.xlabel('regLevel')
plt.ylabel('Level of Overfitting')
plt.legend
```

# Regularisation Tuning

Regularisation restricts the size of the weights, thus the weights are less flexible and cannot fit the training data as well. This however means the model is more likley to follow the trend of the entire population. Thus regularisation can help validation accuracy improve. With too much regularisation the weights are too restriced and they can't fit the training data very well at all and thus don't follow the trend of the population either. Too much regularisation can make the validation accuracy worse.

How to find the sweet spot? Simply by training with multiple regularisation values and plotting regularisation vs accuracy we can find the peak accuracy and the best regularisation parameter.
