# Investigation with Plots

The most obvious way to investigate which attribute has the most affect on the delay time is to plot the relationship between delay time and the attributes. With delay time on the y axis and the BBH attribute ont the x axis. A flat relationship indicates the attribute has little affect on the time delay. A plot with steep gradients indicates the attribute strongly affects the time delay. The inital plots focus on one attribute at a time. For a few attributes that are identified to strongly affect time delay a deeper investigation will be caried out. Another plot is made to show the distribution of the BBH over each attribute. It is important to know how many black holes are simulated with each value of the attribute. For example, a black hole with a high time delay is more likely to have a high time delay because of an attribute value that is more abundant. Also the machine learning techniques deployed later are dependant on these distributions. In regression problems machine learning algorythems are able to make predictions more accuratly for values in which it has a lot of data. This can lead to one attribute being construde as more important because it has a more even spread of black holes across its possible range. This is addressed in the relevant chapter.

**Deeper Investigation**  
The deeper investigation will attempt to plot multiple attributes on a single graph. This will allow for an easier physical interpretation of the features. It will hopefully help highlight the contrast between the attributes in terms of the strength of their affect on the time delay.

#### Data 

* Column 0: identifier of the binary

* Column 8: orbital eccentricity of the binary system at the formation of the second-born compact object.

```python
#unit dictionary
units = {
    'ZAMS' : 'Msun',
    'comp' : 'Msun',
    'merge' : 'Msun',
    't_delay' : 'Gyr',
    'smj_axis' : 'Rsun',
}

#you must select desired columns using the column names in a list. 
#you can take a random subset by selecting the number of BBH you want or the fraction of the total data
#set the default directory to where you have the parquet file
def getData(col = ['m_ZAMS_1', 'm_ZMAS_2', 'm_comp_1', 'm_comp_2', 'm_merge', 't_delay', 'smj_axis', 'alpha', 'f_MT', 'Z','orb_ecc']
            , rRows = 0, dataFraction = 1, dataDir = '~/LCPB/all_BHBs.parquet',seed = None, floats=True):
    if (dataFraction != 1):
        rRows = int(dataFraction*87756770)
    if (rRows == 0):
        return pd.read_parquet(dataDir, columns = col)
    elif (rRows > 0):
        rand = np.random.default_rng(seed = seed)
        rows_selected = rand.integers(low = 0, high = 87756770, size = rRows)
        concf = pd.DataFrame()
        for col_i in col:
            onecol = pd.read_parquet(dataDir, columns = [col_i]).iloc[rows_selected]
            concf = pd.concat([concf, onecol], axis=1)
            del(onecol)
            gc.collect()
        if floats: concf.applymap(lambda element: float(element))
        return concf

#examples
#df = getData(dataFraction=0.1,seed = 12345) #returns all columns but only a random 10% of the rows, with an inputted seed.
#df2 = getData(col = ['m_ZAMS_1','m_ZAMS_2','orb_ecc','m_merge','t_delay'],rRows=5)#returns two selected colums and only 5 random binary systems
```

#### Plotting Code
The following cell holds the functions used to make the various plots

```python
def plotFrequency(df,prop,logx=False,logy=False, nbins=500):
    figalot = plt.figure(figsize=(15,10))
    plt.hist(df[prop], bins = nbins, log=logy, density=True)
    plt.title('BBH Distribution, '+str(nbins)+' bins')
    if logx: plt.xscale('log')
    plt.xlabel(prop)
    plt.ylabel('Probability Density')

def plotMedian(df,prop,xlab = None,logx=False,logy=False,ntks=10):
    def lowerQuartile(stuff):
        return np.quantile(stuff,0.25)
    def upperQuartile(stuff):
        return np.quantile(stuff,0.75)

    med = df.groupby(prop)['t_delay'].agg('median').reset_index()
    lq = df.groupby(prop)['t_delay'].agg(lowerQuartile).reset_index()
    uq = df.groupby(prop)['t_delay'].agg(upperQuartile).reset_index()
    med.columns = [prop, 'median_t_delay']

    med['lower_quartile'] = lq['t_delay']
    med['upper_quartile'] = uq['t_delay']
    
    figAcc = plt.figure(figsize=(20,10))
    plt.plot(med[prop],med['median_t_delay'],label = 'Median')
    plt.xlabel(prop)
    
    if(xlab != None): plt.xticks(ticks = np.sort(np.unique(med[prop])),lables = xlab)
        
    if logx: plt.xscale('log')    
    plt.ylabel('Time Delay '+units['t_delay'])
    if logy: plt.yscale('log')
    plt.title("BBH Time Delay Relationship")
    plt.fill_between(med[prop], med['lower_quartile'], med['upper_quartile'], alpha=0.4, color='darkorange',label = 'Interquartile Range')
    plt.legend()
    return med
    
def qplotMedianMass(df,prop,nbins = 1000,logy=False,logx=False,figwidth=20):
    # group the masses into quantiles and get the bins
    out, bins = pd.qcut(df[prop], q = nbins, retbins = True)
    bhperbin = int(len(df)/nbins)
    prop = prop +', quantiles'
    # add the quantiles to the data frame
    df[prop] = out
    df = df.sort_values(by=prop,ascending=True)
    med = df.groupby(prop)['t_delay'].agg('median').reset_index()
    lq = df.groupby(prop)['t_delay'].quantile(0.25).reset_index()
    uq = df.groupby(prop)['t_delay'].quantile(0.75).reset_index()
    nbbh = df.groupby([prop]).count().reset_index()
    med.columns = [prop, 'median_t_delay']
    
    med['lower_quartile'] = lq['t_delay']
    med['upper_quartile'] = uq['t_delay']
    med['mids'] = df[prop].unique().map(lambda interval: interval.mid)
    figAcc = plt.figure(figsize = (figwidth,10))
    plt.plot(med['mids'],med['median_t_delay'],label = 'Median', marker = '.')
    plt.xlabel(prop)
    if logx: plt.xscale('log')        
    plt.ylabel('Time Delay '+units['t_delay'])
    if logy: plt.yscale('log')
    plt.title('BBH Time Delay Relationship\n'+str(nbins)+' bins with '+str(bhperbin)+' BBH per bin')
    plt.fill_between(med['mids'], med['lower_quartile'], med['upper_quartile'], alpha=0.4,              color='darkorange',label = 'Interquartile Range')
    plt.legend()
    return med
```

## First Look Semi-Majour Axis
This is with regard to the objects eliptical orbit. It is the distance from the center of mass of the orbit to the farthest point of the orbit. The large this is the further the objects are apart when the binary star system is formed. The farther apart the objects are the longer it should take for them to merge and the time delay should be larger.

```python
df = getData(['smj_axis','t_delay'])
prop = 'Semi-Major Axis, '+units['smj_axis']
df.columns = [prop,'t_delay']
```

```python
# plotFrequency(df,prop)
#plt.hist(df[prop])
def plotFrequency2(df,prop,xlab,nbins = 1000, logy=False, logx=False):
    quantile, bins = pd.cut(df[prop], bins=nbins, retbins = True)
    df['quantile'] = quantile
    bin1 = df['quantile'].unique()[0]
    newbin1 = pd.Interval(left=0, right=bin1.right, closed='right')
    df = df.replace(to_replace=bin1,value=newbin1)
    freq = df.groupby('quantile').agg('count').reset_index()
    freq = freq.iloc[:,0:2]
    freq.columns = ['quantile','frequency']
    freq['mids'] = freq['quantile'].map(lambda interval: interval.mid)
    freq['width'] = freq['quantile'].map(lambda interval: interval.length)
    binwidth = np.mean(freq['width'])
    
    figorbec = plt.figure()
    pdf = freq['frequency'] / (np.sum(freq['frequency'])*np.diff(bins))
    plt.bar(freq['mids'],pdf, width=binwidth)
    plt.title('BBH Distribution, '+str(nbins)+' bins')
    if logy: plt.yscale('log')
    if logx: plt.xscale('log')
    plt.xlabel(prop)
    plt.ylabel('Frequency')
    return freq

plotFrequency2(df,prop,logy=True,logx=True, xlab=[10**n for n in (-2,-1,0,1,2)])
```

```python
med = qplotMedianMass(df,prop,nbins=10000,logx=True,figwidth=40)
```

```python
max_med_pos = np.argmax(med['median_t_delay'])
smj_max_med = np.max(med['median_t_delay'])
smj_val_maxmed = med['mids'][max_med_pos]
print('The max median time delay is:', smj_max_med,'Gyr')
print('This occurs at a semi majour axis value of',smj_val_maxmed,'Rsun')
```

#### Discussion
The semi majour axis is essentially a measure of how far apart the objects are in their eliptical orbit around eachother. It is obvious that when the objects are further apart the time delay for their merging is longer. The plot above clearly shows this relationship. Although it was expected to be monotomically increasing. There is a section just below $20 Rsun$ that shows a decrese of time delay with increasing semi majour axis. Close to $10^2 Rsun$ the trend reverses and flat lines at about $2 Gyr$. Objects that bengin very far apart $10^7 Rsun$ can have a time delay similar to that of objects that are $10 or 20 Rsun$. This is strange. Perhaps these BBH are ones with a low common envalope efficiancy which allows the common envalope to bring them drastically closer together. Perhaps they also have the ideal masses for low time delay.

```python
from celluloid import Camera # getting the camera
from IPython.display import HTML # to show the animation in Jupyter
lable2name = {'alpha':'Common Envalope Efficiency','Z':'Metalicity, % of mass','m_ZAMS_1':'ZAMS of Primary object, Msun','m_comp_1':'Mass of Primary Compact Object, Msun','smj_axis':'Semi Majour Axis, Rsun','t_delay':'Time Delay Gyr'}
plt.rc('font',size=10)
colors = ['blue','red','pink','brown','yellow','black','orange','purple','magenta','lime','cyan','green']
zamsP_mass_minmed = 37.528499999999994
compP_mass_minmed = 11.1305
smj_val_maxmed = 62.966
```

```python
df = getData(col=['alpha','Z','m_ZAMS_1','m_comp_1','smj_axis','t_delay'],dataFraction=0.5, floats=False)

df['m_ZAMS_1_groups'] = df['m_ZAMS_1'] < zamsP_mass_minmed
df['m_ZAMS_1_groups'] = df['m_ZAMS_1_groups'].replace(True,'ZAMS Primary Before Minimum: 38 Msun')
df['m_ZAMS_1_groups'] = df['m_ZAMS_1_groups'].replace(False,'ZAMS Primary After Minimum: 38 Msun')

df['m_comp_groups'] = df['m_comp_1'] < compP_mass_minmed
df['m_comp_groups'] = df['m_comp_groups'].replace(True,'Compact Mass Before Minimum: 11 Msun')
df['m_comp_groups'] = df['m_comp_groups'].replace(False,'Compact Mass After Minimum: 11 Msun')

df['smj_groups'] = df['smj_axis'] < smj_val_maxmed
df['smj_groups'] = df['smj_groups'].replace(True,'Semi-Major Axis Before Maximum: 63 Rsun')
df['smj_groups'] = df['smj_groups'].replace(False,'Semi-Major Axis After Maximum: 63 Rsun')
```

```python
def plotSimParamVsMass(df, param_main):
    comp_med = df.groupby([param_main,'m_comp_groups']).median()['t_delay'].unstack()
    comp_lq = df.groupby([param_main,'m_comp_groups']).quantile(0.25)['t_delay'].unstack()
    comp_uq = df.groupby([param_main,'m_comp_groups']).quantile(0.75)['t_delay'].unstack()

    zams_med = df.groupby([param_main,'m_ZAMS_1_groups']).median()['t_delay'].unstack()
    zams_lq = df.groupby([param_main,'m_ZAMS_1_groups']).quantile(0.25)['t_delay'].unstack()
    zams_uq = df.groupby([param_main,'m_ZAMS_1_groups']).quantile(0.75)['t_delay'].unstack()

    smj_med = df.groupby([param_main,'smj_groups']).median()['t_delay'].unstack()
    smj_lq = df.groupby([param_main,'smj_groups']).quantile(0.25)['t_delay'].unstack()
    smj_uq = df.groupby([param_main,'smj_groups']).quantile(0.75)['t_delay'].unstack()
    
    figMass, ax = plt.subplots()
    camera = Camera(figMass)# the camera gets the fig we'll plot
    legend_pos = (0.03,0.97)
    ax.legend(title='BBH Groups',bbox_to_anchor=legend_pos, loc='upper left', ncol=3)
    ax.set_title('BBH Time Delay Relationships\n Split into Groups')
    ax.set_ylabel("Time Delay Gyr")
    tkloc = df[param_main].unique()
    tklbl = range(0,len(tkloc))
    ax.set_xticks(tklbl,tkloc)
    
    comp_colors = colors[0:2]
    zams_colors = colors[2:4]
    smj_colors = colors[4:6]

    comp_med.plot(ax=ax,color=comp_colors)
    zams_med.plot(ax=ax,color=zams_colors)
    smj_med.plot(ax=ax,color=smj_colors)

    handles, labels = ax.get_legend_handles_labels()

    for column, colour in zip(comp_lq.columns,comp_colors):
        comp_med.plot(ax=ax,color=comp_colors)
        zams_med.plot(ax=ax,color=zams_colors)
        smj_med.plot(ax=ax,color=smj_colors)
        ax.fill_between(comp_lq.index, comp_lq[column], comp_uq[column], facecolor = colour, alpha=0.3)
        ax.legend(handles,labels,title='BBH Groups',bbox_to_anchor=legend_pos, loc='upper left', ncol=3)
        camera.snap()

    for column, colour in zip(zams_lq.columns,zams_colors):
        comp_med.plot(ax=ax,color=comp_colors)
        zams_med.plot(ax=ax,color=zams_colors)
        smj_med.plot(ax=ax,color=smj_colors)
        ax.fill_between(zams_lq.index, zams_lq[column], zams_uq[column], facecolor = colour, alpha=0.3)
        ax.legend(handles,labels,title='BBH Groups',bbox_to_anchor=legend_pos, loc='upper left', ncol=3)
        camera.snap()

    for column, colour in zip(smj_lq.columns,smj_colors):
        comp_med.plot(ax=ax,color=comp_colors)
        zams_med.plot(ax=ax,color=zams_colors)
        smj_med.plot(ax=ax,color=smj_colors)
        ax.fill_between(smj_lq.index, smj_lq[column], smj_uq[column], facecolor = colour, alpha=0.3)
        ax.legend(handles,labels,title='BBH Groups',bbox_to_anchor=legend_pos, loc='upper left', ncol=3)
        camera.snap()
    
    ax.set_xlabel(lable2name[param_main])
    if logx: ax.set_xscale('log')
    if logy: ax.set_yscale('log')
    plt.close()
    return camera.animate(interval=2000)
```

```python
def plotSimParams(df,main_param):
    if main_param=='alpha': 
        gbcols = ['alpha','Z']
        other_param = 'Z'
    elif main_param=='Z': 
        gbcols = ['Z','alpha']
        other_param = 'alpha'
    med = df.groupby(gbcols).median()['t_delay'].unstack()
    lowq = df.groupby(gbcols).quantile(0.25)['t_delay'].unstack()
    upq = df.groupby(gbcols).quantile(0.75)['t_delay'].unstack()

    figSim, ax = plt.subplots()
    camera = Camera(figSim)# the camera gets the fig we'll plot

    ax.set_title('BBH Time Delay Relationships\nSplit into Groups')
    ax.set_ylabel('Time Delay Gy')
    tkloc = df[main_param].unique()
    tklbl = range(0,len(tkloc))
    ax.set_xticks(tklbl,tkloc)
    
    med.plot(ax=ax,color=colors)
    ax.legend(title=lable2name[other_param],bbox_to_anchor=(0.9,0.9), loc='lower right', ncol=len(lowq.columns))
    handles, labels = ax.get_legend_handles_labels()
    
    for column, colour in zip(lowq.columns,colors[0:len(lowq.columns)]):
        med.plot(ax=ax,color=colors)
        ax.legend(handles, labels, title=lable2name[other_param],bbox_to_anchor=(0.03,0.97), loc='upper left', ncol=int(len(lowq.columns)/2))
        ax.fill_between(lowq.index,lowq[column],upq[column],facecolor=colour,alpha=0.3)
        if logx: ax.set_xscale('log')
        if logy: ax.set_yscale('log')
        camera.snap() # the camera takes a snapshot of the plot

    ax.set_xlabel(lable2name[main_param])
    if logx: ax.set_xscale('log')
    if logy: ax.set_yscale('log')
    plt.close()
    return camera.animate(interval=2000) # animation ready
```

```python
def qplotMedianMass2(df,prop1,prop2='Z',nbins = 40,logy=False,logx=False):
    # group the masses into quantiles and get the bins
    out, bins = pd.qcut(df[prop1], q = nbins, retbins = True)
    bhperbin = int(len(df)/nbins)
#     prop = prop +', quantiles'
    # add the quantiles to the data frame
    df['quantiles'] = out
    df['mids'] = df['quantiles'].apply(lambda interval: interval.mid)
    figMedMass2, ax = plt.subplots()
    camera = Camera(figMedMass2)
    med = df.groupby(['mids',prop2]).median()['t_delay'].unstack()
    lq = df.groupby(['mids',prop2]).quantile(0.25)['t_delay'].unstack()
    uq = df.groupby(['mids',prop2]).quantile(0.75)['t_delay'].unstack()
    
    for column,colour in zip(med.columns,colors[0:len(lq.columns)]):
        ax.plot(med.index,med[column],color=colour,label=column)
    handles, labels = ax.get_legend_handles_labels()
    for column, colour in zip(lq.columns,colors[0:len(lq.columns)]):
        for column_j,colour_j in zip(med.columns,colors[0:len(lq.columns)]):
            ax.plot(med.index,med[column_j],color=colour_j)
        ax.fill_between(med.index,lq[column],uq[column],facecolor = colour, alpha=0.3)
        ax.legend(handles,labels,title=lable2name[prop2],ncol=int(len(lq.columns)/2))
        if logx: ax.set_xscale('log')
        if logy: ax.set_yscale('log')
        camera.snap()
    ax.set_xlabel(lable2name[prop1])
    ax.set_ylabel(lable2name['t_delay'])
    ax.set_title('Time Delay Relationships\nSplit into Groups')
    if logx: ax.set_xscale('log')
    if logy: ax.set_yscale('log')
    plt.close()
    return camera.animate(interval=2000)
```

```python
def qplotMedian2Mass(df,mass1,mass2,nbins1 = 500,nbins2=5,logy=False,logx=False):
    # group the masses into quantiles and get the bins
    out1, bins1 = pd.qcut(df[mass1], q = nbins1, retbins = True)
    out2, bins2 = pd.qcut(df[mass2], q = nbins2, retbins = True)    
    out2 = np.asarray(out2)
    bhperbin1 = int(len(df)/nbins1)
    bhperbin2 = int(len(df)/nbins2)
    # add the quantiles to the data frame
    df['quantiles1'] = out1
    df['quantiles2'] = out2
    if mass2=='smj_axis':
        beforeMax = pd.Interval(left=31.032, right=62.996)
        afterMax = pd.Interval(left=62.996, right=np.max(df['smj_axis']))
        df.loc[df['smj_axis']>62.996, 'quantiles2'] = afterMax
        df.loc[(df['smj_axis']>31.032) & (df['smj_axis']<62.996), 'quantiles2'] = beforeMax
        
    df['mids1'] = df['quantiles1'].apply(lambda interval: interval.mid)
    df['mids2'] = df['quantiles2'].apply(lambda interval: interval.mid)
    
    figMedMass2, ax = plt.subplots()
    camera=Camera(figMedMass2)
    med = df.groupby(['mids1','quantiles2']).median()['t_delay'].unstack()
    lq = df.groupby(['mids1','mids2']).quantile(0.25)['t_delay'].unstack()
    uq = df.groupby(['mids1','mids2']).quantile(0.75)['t_delay'].unstack()
    for column,colour in zip(med.columns,colors[0:len(lq.columns)]):
        plt.plot(med.index,med[column],color=colour,label=column)
    handles, labels = ax.get_legend_handles_labels()
    for column, colour in zip(lq.columns,colors[0:len(lq.columns)]):
        for column_j,colour_j in zip(med.columns,colors[0:len(lq.columns)]):
            plt.plot(med.index,med[column_j],color=colour_j,label=column_j)
        ax.fill_between(med.index,lq[column],uq[column],facecolor = colour, alpha=0.3)
        ax.legend(handles,labels,title=lable2name[mass2],ncol=int(len(lq.columns)/2))
        if logx: ax.set_xscale('log')
        if logy: ax.set_yscale('log')
        camera.snap()
    ax.set_xlabel(lable2name[mass1])
    ax.set_ylabel(lable2name['t_delay'])
    ax.set_title('Time Delay Relationships\nSplit into Groups')
    if logx: ax.set_xscale('log')
    if logy: ax.set_yscale('log')
    plt.close()
    return camera.animate(interval=2000)
```
