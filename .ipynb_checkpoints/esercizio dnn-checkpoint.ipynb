```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense, Dropout
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split as TTsplit
plt.rcParams['font.size'] = 14
```

```python
# size of each sample
L=2
# random seed for reproducibility
np.random.seed(12345)

%run nonlinear_function.py
```

### Useful functions

```python
from tensorflow.keras import initializers
def create_DNN(act='relu', hidden_layers=0, drop_perc=0.2,
              init=initializers.RandomNormal(mean=0.0, stddev=0.05, seed=None)):
    model = Sequential()
    model.add(Dense(L,input_shape=(L,),activation=act, kernel_initializer=init))
    model.add(Dense(20,activation=act, kernel_initializer=init))
    model.add(Dense(20,activation=act, kernel_initializer=init))
    for i in range(hidden_layers):
        model.add(Dense(10,activation=act, kernel_initializer=init))
    model.add(Dropout(drop_perc))
    model.add(Dense(1,activation=act, kernel_initializer=init))
    return model
def DNN(optimizer='adam', loss='binary_crossentropy',
       act='relu', hidden_layers=0, drop_perc=0.2,
              init=initializers.RandomNormal(mean=0.0, stddev=0.05, seed=None)):
    #this function creates the DNN model and compiles it
    model = create_DNN(act, hidden_layers, drop_perc,init)
    model.compile(loss=loss,
              optimizer=optimizer,
              metrics=['accuracy'])
    return model

def Rescale(x):
    #return (x-x.mean())/np.sqrt(x.var())
    return x/50
def boundaries():    
    ax.plot((-20,-20),(-40,50),c="w")
    ax.plot((-20,50),(-40,-40),c="w")
    ax.plot((-10,50),(50,-10),c="w")
```

### 1.a - 1.b

```python
N_array = np.array([2500,4000,5500])
N_lung = len(N_array)
nepoch = 100
TYPE=1

# stampo N_lung finestre dove plotto i dati
fig,AX=plt.subplots(N_lung,3,figsize=(18,6*N_lung))
for j,N in zip(range(N_lung),N_array):
    x,y = (np.random.random((N,2))-0.5)*100, np.zeros(N)
    for i in range(N):
        # label data according to a nonlinear function "f"
        y[i] = f(x[i],TYPE)
    ax=AX[j,0]
    ax.scatter(x[:,0],x[:,1],c=y)
    ax.set_title('N='+str(N))
    
    x = Rescale(x)
    x_train, x_valid, y_train, y_valid = TTsplit(x, y, test_size=0.2)
    
    model = DNN()
    fit = model.fit(x_train, y_train,
               epochs = nepoch, batch_size = 50,
               validation_data=(x_valid,y_valid),
               verbose=2)
    #Accuracy and Lost plots
    ax=AX[j,1]
    ax.plot(fit.history['accuracy'],label="train")
    ax.plot(fit.history['val_accuracy'],label="valid.")
    ax.set_xlabel('epoch')
    ax.set_ylabel("Accuracy")
    ax.set_ylim(0, 1)
    ax=AX[j,2]
    ax.plot(fit.history['loss'],label="train")
    ax.plot(fit.history['val_loss'],label="valid.")
    ax.set_xlabel('epoch')
    ax.set_ylabel("Loss")
    ax.set_ylim(0, 1)
    ax.legend()
```

GridSearch involved the following parameters: epochs, optimizer, loss function, #hidden_layers, activation function and perc of the dropout

```python
N=2000

# recreating datas
x,y = (np.random.random((N,2))-0.5)*100, np.zeros(N)
for i in range(N):
    # label data according to a nonlinear function "f"
    y[i] = f(x[i],TYPE)
x = Rescale(x)
x_train, x_valid, y_train, y_valid = TTsplit(x, y, test_size=0.2)
print(len(x_train))
```

```python
from sklearn.model_selection import GridSearchCV
from keras.wrappers.scikit_learn import KerasClassifier

# call Keras scikit wrapper
model_gridsearch = KerasClassifier(build_fn=DNN, 
                        epochs=1, 
                        batch_size=50, 
                        verbose=1)

# list of allowed optional arguments for the optimizer, see `compile_model()`
epochs_list = [1, 10, 50]
optimizer = ['SGD', 'RMSprop', 'Adagrad', 'Adadelta', 'Adam', 'Adamax', 'Nadam']
losses = ['mean_squared_error', 'squared_hinge', 'binary_crossentropy', 'kl_divergence']
act_functions = ['relu', 'sigmoid', 'elu', 'softsign']
hidden_layers = [0, 2, 4]
drop_percs = [0, 0.2, 0.4]
parameters = {'loss' : losses}

# call scikit grid search module
grid = GridSearchCV(estimator=model_gridsearch, param_grid=parameters, n_jobs=1, cv=4)
grid_result = grid.fit(x_train,y_train)

# summarize results
print("Best score: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print("%f (%f) with: %r" % (mean, stdev, param))
```

## 3
Rescaling data: Normalization and Standardization

```python
# data normalization
from sklearn.preprocessing import MinMaxScaler
# data standardization
from sklearn.preprocessing import StandardScaler
scalers = [MinMaxScaler(), StandardScaler()]
scalers_name = ['MinMaxScaler', 'StandardScaler']
N=4000
nepoch = 400
fig,AX=plt.subplots(len(scalers),2,figsize=(18,6*len(scalers)))
for (j,scaler) in zip(range(len(scalers)),scalers):
    model = DNN()
    # recreating datas
    x,y = (np.random.random((N,2))-0.5)*100, np.zeros(N)
    for i in range(N):
        # label data according to a nonlinear function "f"
        y[i] = f(x[i],TYPE)
    scaler.fit(x)
    x = scaler.transform(x)
    x_train, x_valid, y_train, y_valid = TTsplit(x, y, test_size=0.2)
    #fit
    fit = model.fit(x_train, y_train,
               epochs = nepoch, batch_size = 50,
               validation_data=(x_valid,y_valid),
               verbose=2)

    #Accuracy and Lost plots
    ax=AX[j,0]
    ax.plot(fit.history['accuracy'],label="train")
    ax.plot(fit.history['val_accuracy'],label="valid.")
    ax.set_xlabel('epoch')
    ax.set_ylabel("Accuracy")
    ax.set_ylim(0, 1)
    ax.set_title('scaler='+scalers_name[j]+' ACCURACY')
    ax=AX[j,1]
    ax.plot(fit.history['loss'],label="train")
    ax.plot(fit.history['val_loss'],label="valid.")
    ax.set_xlabel('epoch')
    ax.set_ylabel("Loss")
    ax.set_ylim(0, 1)
    ax.set_title('scaler='+scalers_name[j]+' LOSS')
    ax.legend()
```

below we analyze the DNN performance for different weight initializers

```python
initializers_list = [initializers.Zeros(),    #all zeros
                     initializers.RandomNormal(mean=0.0, stddev=0.05, seed=None),       #random with normal distr
                     initializers.TruncatedNormal(mean=0.0, stddev=0.05, seed=None),    #rnd with truncated norm distr
                     initializers.Ones()]     #all ones
```

```python
# call Keras scikit wrapper
mdl_gridsearch = KerasClassifier(build_fn=DNN, 
                        epochs=1, 
                        batch_size=50, 
                        verbose=1)
parameters = {'init' : initializers_list}
# call scikit grid search module
grid = GridSearchCV(estimator=mdl_gridsearch, param_grid=parameters, n_jobs=1, cv=4)
grid_result = grid.fit(x_train,y_train)
# summarize results
print("Best score: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print("%f (%f) with: %r" % (mean, stdev, param))
```
