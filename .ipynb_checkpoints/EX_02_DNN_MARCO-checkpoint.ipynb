```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Dropout
import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 14

%run nonlinear_function.py

# training data
# Keras works with numpy arrays: just use them from the start
TYPE=1
x = np.loadtxt(filename("data",TYPE), delimiter=' ')
y = np.loadtxt(filename("labels",TYPE), delimiter=' ')
N = len(x)

# dim. of a sample
L = len(x[0])

plt.figure(figsize = (8,8))
plt.scatter(x[:,0],x[:,1],c=y)
plt.show
```

### Define the model with Keras

```python
model = Sequential()
model.add(Dense(L,input_shape=(L,),activation='relu'))
model.add(Dense(20,activation='relu'))
model.add(Dense(20,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1,activation='sigmoid'))

print(model.summary())
#optimizer
model.compile(loss='binary_crossentropy',
              #optimizer='RMSprop',
              optimizer='adam',
              metrics=['accuracy'])
```

## Rescaling the datas

```python
def Rescale(x):
    #return (x-x.mean())/np.sqrt(x.var())
    return x/50

x = Rescale(x)
```

## 1.a - 1.b

```python
#defining a set of perc_train: 10 percents from 65% to 95%
percents = np.linspace(0.65, 0.95, 4)
print(percents)
nepoch=400

from keras import backend as K
from sklearn.model_selection import train_test_split as TTsplit

fig,AX=plt.subplots(len(percents),2,figsize=(14,8*(len(percents))))

for i,perc_train in zip(range(len(percents)),percents):
    x_train, x_valid, y_train, y_valid = TTsplit(x, y, test_size=perc_train)
    fit = model.fit(x_train, y_train,
               epochs = nepoch, batch_size = 50,
               validation_data=(x_valid,y_valid),
               verbose=2)
    #Accuracy and Lost plots
    ax=AX[i,0]
    ax.plot(fit.history['accuracy'],label="train")
    ax.plot(fit.history['val_accuracy'],label="valid.")
    ax.set_xlabel('epoch')
    ax.set_ylabel("Accuracy")
    ax.set_ylim(0, 1)
    ax=AX[i,1]
    ax.plot(fit.history['loss'],label="train")
    ax.plot(fit.history['val_loss'],label="valid.")
    ax.set_xlabel('epoch')
    ax.set_ylabel("Loss")
    ax.set_ylim(0, 1)
    ax.legend()
    
    K.clear_session()
```
